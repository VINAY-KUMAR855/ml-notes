```python
l = [5,6,3,2,8,7]
def sort(l):
    for i in range(len(l)):
        min = i
        for j in range(i+1,len(l)):
            if l[min]>l[j]:
                min = j
        l[i],l[min] = l[min],l[i]
    print(l)
sort(l)
```
