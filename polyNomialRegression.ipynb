```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/majji-vinaykumar/Videos/mechile_Learning/Ice_cream selling data.csv")
df.head()
```

```python
x = df.iloc[:,0:1].values
y = df.iloc[:,1].values
```

```python
print(x.shape)
print(x)
```

```python
print(y.shape)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.intercept_  # teta0 or c value
```

```python
model.coef_  #teta1 value
```

```python
model.score(x_train,y_train)
#if valus is low then it is underfitting
```

```python
model.score(x_test,y_test)
```

```python
pred = model.predict(x_train)
```

```python
pd.DataFrame([x_train.reshape(-1),y_train,model.predict(x_train)]).T
```

```python
plt.scatter(x_train,y_train,c="blue")
plt.plot(x_train,model.predict(x_train),c="blue")
plt.show()
```

```python
plt.scatter(x_test,y_test,c="blue")
plt.plot(x_test,model.predict(x_test),c="blue")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
degree =2
poly = PolynomialFeatures(degree)
```

```python
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.fit_transform(x_test)
```

```python
x_train_poly
```

```python
x_test_poly
```

```python
model = LinearRegression()
model.fit(x_train_poly,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(x_train_poly,y_train)
```

```python
model.score(x_test_poly,y_test)
```

```python
predection = model.predict(x_test_poly)
```

```python
#original points
plt.scatter(x_train,y_train,color="red",label="actual")
#create a smooth curve for plotting
x_range= np.linspace(x_train.min(),x_train.max(),100).reshape(-1,1)# more points = smooth curve
x_range_poly = poly.transform(x_range)# transform using same polynomial features
#pradict using trained model
y_pred_curve = model.predict(x_range_poly)
#plotting the polynomial regression curve
plt.plot(x_range,y_pred_curve,label="polynomial Regression curve")
plt.xlabel("Temperature or x")
plt.ylabel("Ice cream sales or y")
plt.legend()
plt.show()
```
