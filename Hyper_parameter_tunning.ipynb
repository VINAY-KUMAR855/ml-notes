```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
#print(breast_cancer)
```

```python
df = pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
```

```python
df.head()
```

```python
df["label"] = breast_cancer.target
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
df["label"].value_counts()
#1 means benign, 0 means malignant
```

```python
x = df.drop("label",axis=1)
y = df["label"]
```

```python
#print(x)
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

# gridsearchcv

```python
# gridsearchcv is used for determinig the best parameter for our model
```

```python
model = SVC()
```

```python
#hyperperameter
parameter = {
    "kernel":["linear","poly","rbf","sigmoid"],
    "C":[1,5,10,20]
}
```

```python
classifier = GridSearchCV(model,parameter,cv=5)  # cv mearns cross validation
```

```python
# fitting the data to our model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
# best parameters
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# highest accuracy for above best parameter
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
#loading results in dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_result = result[["param_C","param_kernel","mean_test_score"]]
```

```python
grid_result
```
