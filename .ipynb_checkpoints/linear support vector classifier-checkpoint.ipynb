```python
import numpy as np
from sklearn import datasets 
```

```python
iris = datasets.load_iris()
```

```python
iris
```

'feature_names': ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], \
  'target_names': array(['setosa', 'versicolor', 'virginica']

```python
x = iris['data'][:,(2,3)] # petal length, petal width
y = (iris['target']==2).astype(np.float64)
```

```python
#(iris['target']==2)
```

```python
x[:5]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
svm_clf = Pipeline([
    ('scaler',StandardScaler()),
    ('linear_svc',LinearSVC(C=1,loss='hinge')),
])
```

```python
svm_clf.fit(x,y)
```

```python
svm_clf.predict([[5.5, 1.7]])
```

```python
# below code is from gemini
```

```python
# 1. Define the grid of points to evaluate (The "canvas" for the boundary)
# We find the min/max of the data and extend slightly
x0_min, x0_max = x[:, 0].min() - 0.5, x[:, 0].max() + 0.5
x1_min, x1_max = x[:, 1].min() - 0.5, x[:, 1].max() + 0.5

# Create a meshgrid (a dense grid of all possible (x, y) combinations)
x0, x1 = np.meshgrid(
    np.linspace(x0_min, x0_max, 500),
    np.linspace(x1_min, x1_max, 500)
)
# Flatten the grid into a list of points (like X)
X_new = np.c_[x0.ravel(), x1.ravel()]

# 2. Make predictions on every point in the grid
y_predict = svm_clf.predict(X_new)

# 3. Reshape the predictions back into a grid format
zz = y_predict.reshape(x0.shape)

# 4. Plot the results
plt.figure(figsize=(10, 7))

# Plot the decision boundary and the margin
# contourf fills the regions with color based on prediction (0 or 1)
plt.contourf(x0, x1, zz, cmap=plt.cm.Spectral, alpha=0.3)
plt.contour(x0, x1, zz, colors='k', linewidths=2, alpha=0.9) # This plots the boundary line

# Plot the original data points on top
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Spectral)

# --- Formatting ---
plt.title('Linear SVM Decision Boundary on Iris-Virginica Data')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.xlim(x0_min, x0_max)
plt.ylim(x1_min, x1_max)
plt.legend(handles=[
    plt.Line2D([0], [0], marker='o', color='w', label='Not Virginica (0.0)',
               markerfacecolor='blue', markersize=10),
    plt.Line2D([0], [0], marker='o', color='w', label='Iris-Virginica (1.0)',
               markerfacecolor='red', markersize=10)
])
plt.show()
```
