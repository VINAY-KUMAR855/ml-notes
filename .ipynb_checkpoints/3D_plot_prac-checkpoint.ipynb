```python
# Step 1: Create arrays for x and y ranges
x = np.array([1, 2, 3])  # 3 x-values
y = np.array([10, 20])   # 2 y-values

print("x values:", x)
print("y values:", y)

# Step 2: Create meshgrid - makes ALL combinations!
X, Y = np.meshgrid(x, y)
```

```python

plt.scatter(X,Y)

plt.show()
```

```python
# Create a spiral path (like a bird circling upward)
t = np.linspace(0, 10, 100)  # time from 0 to 10 seconds

# Position changes over time
x = np.cos(t)      # East-West (circular motion)
y = np.sin(t)      # North-South (circular motion)
z = t              # Height (going up!)

# Create 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the line (connects the points)
ax.plot(x, y, z, 'b-', linewidth=2, label='Bird Flight Path')

# Mark start and end
ax.scatter([x[0]], [y[0]], [z[0]], c='green', s=200, marker='o', label='Start')
ax.scatter([x[-1]], [y[-1]], [z[-1]], c='red', s=200, marker='*', label='End')

ax.set_xlabel('X (East-West)')
ax.set_ylabel('Y (North-South)')
ax.set_zlabel('Z (Height)')
ax.set_title('3D Line Plot: Spiral Path', fontsize=14, fontweight='bold')
ax.legend()

plt.show()

print(f"Start position: ({x[0]:.2f}, {y[0]:.2f}, {z[0]:.2f})")
print(f"End position: ({x[-1]:.2f}, {y[-1]:.2f}, {z[-1]:.2f})")
```

```python
# Create a surface
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Ripple pattern!

# Create 4 subplots with different viewing angles
fig = plt.figure(figsize=(14, 10))

# View 1: Top view (bird's eye)
ax1 = fig.add_subplot(221, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='coolwarm')
ax1.view_init(elev=90, azim=0)  # Look straight down
ax1.set_title('Top View (elev=90)', fontweight='bold')

# View 2: Side view
ax2 = fig.add_subplot(222, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='coolwarm')
ax2.view_init(elev=0, azim=0)   # Look from the side
ax2.set_title('Side View (elev=0)', fontweight='bold')

# View 3: Default angle
ax3 = fig.add_subplot(223, projection='3d')
ax3.plot_surface(X, Y, Z, cmap='coolwarm')
ax3.view_init(elev=30, azim=45)  # Nice 3D perspective
ax3.set_title('Angle View (elev=30, azim=45)', fontweight='bold')

# View 4: Another angle
ax4 = fig.add_subplot(224, projection='3d')
ax4.plot_surface(X, Y, Z, cmap='coolwarm')
ax4.view_init(elev=20, azim=135)  # Look from opposite side
ax4.set_title('Different Angle (elev=20, azim=135)', fontweight='bold')

plt.tight_layout()
plt.savefig('viewing_angles.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# The temperature function from calculus lesson
def T(x, y):
    return x**2 + y**2 + 2*x*y

# Partial derivatives (you'll learn what these mean in calculus)
def dT_dx(x, y):
    return 2*x + 2*y

def dT_dy(x, y):
    return 2*y + 2*x

# Create grid - YOU UNDERSTAND MESHGRID NOW!
x = np.linspace(-3, 3, 100)  #  100 x values from -3 to 3
y = np.linspace(-3, 3, 100)  #  100 y values from -3 to 3
X, Y = np.meshgrid(x, y)     #  Creates 100100 = 10,000 points!
Z = T(X, Y)                  #  Calculate temperature at each point

# Create figure - YOU KNOW THIS!
fig = plt.figure(figsize=(15, 5))

# 3D surface plot - YOU CAN DO THIS!
ax1 = fig.add_subplot(131, projection='3d')  #  First of 3 subplots, make it 3D
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)  #  Surface plot!
ax1.set_title('Temperature Function T(x,y)', fontsize=14, fontweight='bold')
ax1.set_xlabel('X (meters East)')
ax1.set_ylabel('Y (meters North)')
ax1.set_zlabel('Temperature')

# Partial derivative plot - SAME TECHNIQUE!
ax2 = fig.add_subplot(132, projection='3d')
dZ_dx = dT_dx(X, Y)  #  How fast temp changes in X direction
ax2.plot_surface(X, Y, dZ_dx, cmap='coolwarm', alpha=0.8)
ax2.set_title('T/x (Rate of change moving East)', fontsize=14, fontweight='bold')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('T/x')

# Another partial derivative - YOU UNDERSTAND EVERY LINE NOW!
ax3 = fig.add_subplot(133, projection='3d')
dZ_dy = dT_dy(X, Y)  #  How fast temp changes in Y direction
ax3.plot_surface(X, Y, dZ_dy, cmap='plasma', alpha=0.8)
ax3.set_title('T/y (Rate of change moving North)', fontsize=14, fontweight='bold')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('T/y')

plt.tight_layout()
plt.show()

print("NOW YOU UNDERSTAND EVERY LINE OF THIS CODE! ")
```
