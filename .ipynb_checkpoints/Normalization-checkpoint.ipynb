```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/majji-vinaykumar/Videos/mechile_Learning/wine.csv",usecols=[0,1,2])
df
```

```python
sns.histplot(df['Alcohol'],kde=True)
plt.show()
```

```python
sns.histplot(df['Malic.acid'],kde=True)
plt.show()
```

```python
sns.scatterplot(x='Alcohol',y="Malic.acid",data=df,hue='Wine',palette={1:'black',2:'red',3:'green'})
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Wine',axis=1),df['Wine'],test_size=0.25,random_state=3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## normalization

```python
from sklearn.preprocessing import MinMaxScaler
scalar= MinMaxScaler()
scalar.fit(x_train)  #find min and max values in Alcohol and malicacid
```

```python
x_train_normalized = scalar.transform(x_train)
x_test_normalized = scalar.transform(x_test)
```

```python
x_train.describe()
```

```python
# x_train_normalized is a array so convert into dataframe to know min,max,describe
x_train_normalized = pd.DataFrame(x_train_normalized)
x_train_normalized.columns = x_train.columns
x_train_normalized.describe()
```

```python
x_test.describe()
```

```python
x_test_normalized = pd.DataFrame(x_test_normalized)
x_test_normalized.columns = x_test.columns
x_test_normalized.describe()
```

```python
# in above why min is negative. because in fit fit method we pass only train data, in train data min and max as not same in test data
```

```python
#check no difference in points but range is changed
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.scatterplot(x=x_train['Alcohol'],y=x_train['Malic.acid'])

plt.subplot(1,2,2)
sns.scatterplot(x=x_train_normalized['Alcohol'],y=x_train_normalized["Malic.acid"])

plt.show()
```

```python
#alcohol and malic acid comes same range in fig2
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(x_train['Alcohol'])
sns.kdeplot(x_train['Malic.acid'])

plt.subplot(1,2,2)
sns.kdeplot(x_train_normalized['Alcohol'])
sns.kdeplot(x_train_normalized['Malic.acid'])

plt.show()
```
