```python
#create feature matrix and dependent variable
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
x
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
x[:,1:3] = imputer.fit_transform(x[:,1:3])
```

```python
x
```

```python
#Nominal data- no order like gender,city,color; we can,t say low or high
#import sklearn
#print(sklearn.__version__)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encode', OneHotEncoder(), [0])], remainder = 'passthrough' )#remainder is for keep other columns
x = ct.fit_transform(x)
#for ordinal data like size, rank, Rating
#from sklearn.preprocessing import LabelEncoder
#label_encoder = LabelEncoder()
#x[:,0] = label_encoder.fit_transform(x[:,0])
```

```python
x
```

```python
x = x[:,1:] #avoiding the dummy variable trap; if two is zero the reain are choose
```

```python
#split into taining set and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```
