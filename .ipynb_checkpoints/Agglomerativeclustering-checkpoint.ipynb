```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/majji-vinaykumar/Documents/Mall_Customers.csv")
df.head()
```

```python
x = df.iloc[:,3:].values
```

```python
# Perform hierarchical clustering
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(x, method='ward') #Creates steps for clustering (linkage matrix)
# Plot dendrogram
plt.figure(figsize=(8, 4))
dendrogram(linked)#Draws the dendrogram/tree using linkage matrix
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python

plt.figure(figsize=(8, 4))
dendrogram(linked)
plt.axhline(y=200, color='r', linestyle='--')
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, linkage='ward')
y_hc = hc.fit_predict(x)
```

```python
df["clusters"] = y_hc
```

```python
df
```

```python
df1 = df[df["clusters"]==0]
df2 = df[df["clusters"]==1]
df3 = df[df["clusters"]==2]
df4 = df[df["clusters"]==3]
df5 = df[df["clusters"]==4]
```

```python
plt.scatter(df1["Annual Income (k$)"],df1["Spending Score (1-100)"],color="red")
plt.scatter(df2["Annual Income (k$)"],df2["Spending Score (1-100)"],color="blue")
plt.scatter(df3["Annual Income (k$)"],df3["Spending Score (1-100)"],color="black")
plt.scatter(df4["Annual Income (k$)"],df4["Spending Score (1-100)"],color="green")
plt.scatter(df5["Annual Income (k$)"],df5["Spending Score (1-100)"],color="pink")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score ")
plt.show()
```
