```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/majji-vinaykumar/Videos/mechile_Learning/Social_Network_Ads.csv",usecols=[2,3,4])
df.head()
```

```python
sns.histplot(df['Age'],kde=True)
```

```python
sns.histplot(df['EstimatedSalary'],kde=True)
```

```python
sns.scatterplot(x="Age",y="EstimatedSalary",data=df,hue='Purchased')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.25,random_state=3)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(x_train)
```

```python
scalar.mean_
```

```python
x_train_standardization = scalar.transform(x_train)
x_test_standardization = scalar.transform(x_test)
```

```python
x_train.describe()
```

```python
x_train_standardization = pd.DataFrame(x_train_standardization)
x_train_standardization.colums = x_train.columns
#x_train_standardization.describe()
np.round(x_train_standardization.describe(),1)
```

```python
x_test_standardization = pd.DataFrame(x_test_standardization)
x_test_standardization.colums = x_test.columns
np.round(x_test_standardization.describe(),1)
```

```python
x_train_standardization
```

```python
#check no difference in points but range is changed
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.scatterplot(x=x_train['Age'],y=x_train['EstimatedSalary'])

plt.subplot(1,2,2)
sns.scatterplot(x=x_train_standardization[0],y=x_train_standardization[1])

plt.show()
```

```python
#Age and EstimatedSalary comes same range in fig2
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(x_train['Age'])
sns.kdeplot(x_train['EstimatedSalary'])

plt.subplot(1,2,2)
sns.kdeplot(x_train_standardization[0])
sns.kdeplot(x_train_standardization[1])

plt.show()
```
