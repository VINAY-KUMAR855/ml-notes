```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a figure 
fig = plt.figure(figsize=(10, 7)) 
# Create a 3D axis - THIS IS THE KEY DIFFERENCE! 
ax = fig.add_subplot(111, projection='3d') # projection='3d' makes it 3D! 
 # Let's plot 5 random points in 3D space 
x = [1, 2, 3, 4, 5] 
y = [2, 3, 1, 4, 2] 
z = [1, 4, 2, 5, 3] # This is NEW - the height! 
# scatter() in 3D takes x, y, AND z 
ax.scatter( x, y, z, c='red', marker='o', s=100 )

# Add labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')  # NEW: Z label!
ax.set_title('My First 3D Plot!')

plt.show()
```

```python

# Data: 5 points
x = [1, 2, 3, 4, 5]
y = [2, 3, 1, 4, 2]
z = [1, 4, 2, 5, 3]

# Create figure with two subplots
fig = plt.figure(figsize=(15, 6))

# ============= LEFT: 2D Plot (What you know) =============
ax1 = fig.add_subplot(121)  # Normal 2D subplot
ax1.scatter(x, y,
           c='blue',
           marker='o',
           s=100)
ax1.set_xlabel('X axis', fontsize=12)
ax1.set_ylabel('Y axis', fontsize=12)
ax1.set_title('2D Plot (Top View)\nWe LOSE height information!',
              fontsize=14,
              fontweight='bold')
ax1.grid(True, alpha=0.3)

# Add text showing we're missing Z
for i in range(len(x)):
    ax1.text(x[i], y[i],
             f'  ({x[i]},{y[i]},?)',
             fontsize=9)

# ============= RIGHT: 3D Plot (New!) =============
ax2 = fig.add_subplot(122, projection='3d')  # 3D subplot!
ax2.scatter(x, y, z,
           c='red',
           marker='o',
           s=100)
ax2.set_xlabel('X axis', fontsize=12)
ax2.set_ylabel('Y axis', fontsize=12)
ax2.set_zlabel('Z axis (HEIGHT)', fontsize=12)  # Z label!
ax2.set_title('3D Plot\nNow we see ALL information!',
              fontsize=14,
              fontweight='bold')

# Add text showing complete coordinates
for i in range(len(x)):
    ax2.text(x[i], y[i], z[i],
             f'  ({x[i]},{y[i]},{z[i]})',
             fontsize=9)

plt.tight_layout()
plt.savefig('2d_vs_3d.png',
            dpi=300,
            bbox_inches='tight')
plt.show()

print("Notice: 2D plot looks down from above (bird's eye view)")
print("        3D plot shows the true position in space!")
```

```python
# Step 1: Create arrays for x and y ranges
x = np.array([1, 2, 3])  # 3 x-values
y = np.array([10, 20])   # 2 y-values

print("x values:", x)
print("y values:", y)

# Step 2: Create meshgrid - makes ALL combinations!
X, Y = np.meshgrid(x, y)

print("\nX grid (all x-coordinates):")
print(X)
# Output: [[1, 2, 3],
#          [1, 2, 3]]

print("\nY grid (all y-coordinates):")
print(Y)
# Output: [[10, 10, 10],
#          [20, 20, 20]]

print("\nThese create points:")
print("(1,10), (2,10), (3,10)")
print("(1,20), (2,20), (3,20)")
print("Total: 6 points in a grid!")
```

```python

# STEP 1: Create x and y ranges
x = np.linspace(-3, 3, 50)  # 50 points from -3 to 3
y = np.linspace(-3, 3, 50)  # 50 points from -3 to 3

# STEP 2: Create meshgrid (the "net")
X, Y = np.meshgrid(x, y)
# Now we have 5050 = 2500 (x,y) points!

# STEP 3: Calculate z for each (x,y) point
Z = X**2 + Y**2  # This works element-wise!

# STEP 4: Create the 3D plot
fig = plt.figure(figsize=(12, 5))

# Left plot: Surface plot
ax1 = fig.add_subplot(121, projection='3d')
surface = ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z = X + Y')
ax1.set_title('3D Surface: z = x + y', fontweight='bold')
fig.colorbar(surface, ax=ax1, shrink=0.5)  # Add color scale

# Right plot: Wireframe (see-through version)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(X, Y, Z, color='blue', alpha=0.6)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z = X + Y')
ax2.set_title('3D Wireframe (Same Function)', fontweight='bold')

plt.tight_layout()
plt.savefig('first_surface.png', dpi=300, bbox_inches='tight')
plt.show()

print("This bowl shape is what a simple loss function looks like!")
print("The minimum is at (0,0,0) - the bottom of the bowl.")
```
