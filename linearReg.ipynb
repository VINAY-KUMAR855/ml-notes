```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/majji-vinaykumar/Videos/mechile_Learning/salary_data.csv")
df.head()
```

```python
x = df.iloc[:,0:1].values
y = df.iloc[:,1].values
print(x.shape, y.shape)
```

```python
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=2)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# y= mx+c
print("c value is",model.intercept_)
print("m value is",model.coef_)
```

```python
model.score(x_train,y_train)#R2 score
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_train)
print(y_train,y_pred)
```

```python
plt.scatter(x_train,y_train,c="blue")
plt.plot(x_train,y_pred,c="blue")
plt.xlabel("experience")
plt.ylabel("salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,c="blue")
plt.plot(x_test,model.predict(x_test),c="blue")
plt.xlabel("experience")
plt.ylabel("salary")
plt.show()
```
