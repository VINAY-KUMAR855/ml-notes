```python
cat_col = ["Department","Role","Industry","Organization","Designation","Last_Appraisal_Rating"]
for col in cat_col:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
"""num_col = ["Passing_Year_Of_Graduation","Passing_Year_Of_PG"]
for col in num_col:
    df[col] = df[col].fillna(df[col].median())"""
df[["Graduation_Specialization","University_Grad","PG_Specialization","University_PG"]]=df[["Graduation_Specialization","University_Grad","PG_Specialization","University_PG"]].fillna("not_studied")
df[["Passing_Year_Of_Graduation","Passing_Year_Of_PG"]]=df[["Passing_Year_Of_Graduation","Passing_Year_Of_PG"]].fillna(0)
```

```python
df.drop(columns=["IDX","Applicant_ID"],inplace=True)
```

```python
df.replace({"Inhand_Offer":{"Y":1,"N":0}},inplace=True)
        
```

```python
df["Inhand_Offer"].value_counts()
```

```python
obj_col = df.select_dtypes(include="object").columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in obj_col:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
