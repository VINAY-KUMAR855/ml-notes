```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_set/housing.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# create a categorical column based on median_income
df["cate_median_income"] = pd.cut(df["median_income"],bins=[0.,1.5,3.0,4.5,6.,np.inf],labels=[1,2,3,4,5])
```

```python
df.head()
```

```python
# stratified shuffle using cate_median_income
```

```python
# we splts data into 10 different times.
# split() method returns indices (row numbers)
# strat_splits.append(...): It takes the training/test pair and saves it into a list. Since n_splits=10, this list will eventually hold 10 different pairs of dataframes.
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
splitter = StratifiedShuffleSplit(n_splits=10,test_size=0.2,random_state=42)
splits = [] # contains all 10 splits, with row 
for train_idx, test_idx in splitter.split(df,df["cate_median_income"]):
    train_val = df.iloc[train_idx]
    test_val = df.iloc[test_idx]
    splits.append([train_val,test_val])
```

```python
train,test = splits[0]
```
