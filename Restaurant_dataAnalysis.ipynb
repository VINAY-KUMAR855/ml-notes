## filling missing values with mode

```python
#find mode
missing = data['Cuisines'].mode()[0]
# fill mode
data['Cuisines'].fillna(missing,inplace=True)
#check
data['Cuisines'].isnull().sum()
```

# data cleaning

```python
# checking for unwanted characters - it's there in categorical columns
```

```python
cat
```

```python
for col in cat:
    data[col] = data[col].str.replace(r'[^A-Za-z0-9-,/]','',regex=True)
```
