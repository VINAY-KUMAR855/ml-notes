```python
# encoding is used to transform sting values into number
#there different types 1)label encoding 2)ordinal encoding 3) onehotencoding
#in label encoding model asign a random value. for example ["hyd","vizag","rajam"] is [1,2,0]
#in ordinal encoding we give vales in specic order.for example ["a","b","c","d"] is [0,1,2,3]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/majji-vinaykumar/Videos/mechile_Learning/DataEncoding.csv")
df.head()
```

```python
#for output column use label encoding. class and grade use ordinal encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Rich','Poor','Middle'],['A','B','C','D']])
oe.fit(df.iloc[:,1:3])
df.iloc[:,1:3]=oe.transform(df.iloc[:,1:3])
```

```python
df
```

```python
oe.categories_
```

```python
oe.inverse_transform(df.iloc[:,1:3])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df.iloc[:,0])
df.iloc[:,0] = le.transform(df.iloc[:,0])
```

```python
df
```

```python
#check order
le.classes_ 
```

```python
le.inverse_transform(list(df.iloc[:,0]))#take list
```
