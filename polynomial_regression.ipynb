```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/home/majji-vinaykumar/Documents/Position_Salaries.csv")
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
print(y)
```

```python
plt.scatter(X,y, color = 'red')
plt.xlabel('position level')
plt.ylabel('Salay')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(X)
```

```python
x_poly
```

```python
#fit model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_poly,y)
```

```python
y_pred = lin_reg.predict(x_poly)
```

```python
plt.scatter(X,y , color="red")
plt.plot(X, y_pred, color = "blue")
plt.title("polynomial regression")
plt.xlabel("Level")
plt.ylabel("salary")
plt.show()
```

```python
#suppose we have to predict salary of a person whose level is 5.5
level = [[5.5]]
level_poly = poly.fit_transform(level)
p_salary = lin_reg.predict(level_poly)
print(p_salary)
```
