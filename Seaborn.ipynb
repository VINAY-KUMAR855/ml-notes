### bar plot

```python
plt.figure(figsize=(10,6),dpi=200)
sns.barplot(x="sepal.length", y="sepal.width",data=iris, estimator=np.mean, errorbar="sd")
plt.show()
#blue color is mean and violet line is st(standard deviation)
```

### pair grid

```python
g= sns.PairGrid(iris)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot,lw=2)
g = g.map_lower(sns.kdeplot,color='red')
```

```python
g = sns.PairGrid(iris, hue="variety",palette="viridis")
g = g.map_upper(sns.scatterplot,linewidths=1,edgecolor='w',s=40)
g= g.map_diag(sns.histplot)
g=g.map_lower(sns.kdeplot)
g = g.add_legend();
```
