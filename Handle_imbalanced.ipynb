```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
sns.counplot(y)
```

## undersampling

```python
from imblearn.under_sampling import NearMiss
```

```python
nm =NearMiss(random_state = 0.2)
x_res, y_res = nm.fit_sample(x,y)
```

```python
print(x.shape)
print(x_res.shape)
print(y_res.shape)
```

```python
from collections import Counter
print(f"Before undersampling:{Counter(y)}")
print(f"Before undersampling:{Counter(y_res)}")
```

## oversampling

### synthetic oversampling

```python
from imblearn.combine import SMOTETomek
sm = SMOTETomek(random_state=0)
x_res, y_res = nm.fit_sample(x,y)
```

```python
print(f"Before undersampling:{Counter(y)}")
print(f"Before undersampling:{Counter(y_res)}")
```

```python
print(x.shape)
print(x_res.shape)
```

### random oversampling

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state = 0.2)
x_res, y_res = nm.fit_sample(x,y)
```

```python
print(f"Before undersampling:{Counter(y)}")
print(f"Before undersampling:{Counter(y_res)}")
```
