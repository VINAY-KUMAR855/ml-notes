```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow
```

```python
class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
```

```python
def main():
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec_())
```

```python
if __name__=="__main__":
    main()
```
